### Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import confusion_matrix,accuracy_score
```

### Importing the dataset 

```python
data = pd.read_csv("C:/Users/Hmz/Downloads/Compressed/Predict_customer_Will_Leave_the-Bank-master/BankCustomers.csv")
```

```python
data
```

### Dependent and independent variables 

```python
X = data.iloc[:, 3:13]
y = data.iloc[:, 13]
```

```python
y
```

### Converting categorical features to dummy variables 

```python
states= pd.get_dummies(X['Geography'])
gender = pd.get_dummies(X['Gender'])
```

```python
states.head(6)
```

```python
#concatenate dummies
X=pd.concat([X,states,gender],axis=1)
```

```python
X
```

```python
#droping the columns as it is no longer required

X = X.drop(['Geography','Gender'],axis=1)
```

```python
X
```

### Splitting the dataset 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Feature Scaling 

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Creating the ANN model 

```python
# Initialising the ANN
classifier = Sequential()

# Adding the second hidden layer
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))

# Adding the output layer
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

### Making the Prediction and evaluating  

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy=accuracy_score(y_test,y_pred)
```
